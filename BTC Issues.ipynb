```python
import pandas as pd
tweets_data_df = pd.read_json("tweets.json", orient = 'records', lines = True)
tweets_data_df.head(5)
tweets_df = pd.DataFrame()
tweets_df['created_at'] = pd.to_datetime(tweets_data_df['created_at'])
tweets_df['text'] = tweets_data_df['text']
tweets_df.index = tweets_df['created_at'] 
del tweets_df['created_at']
tweets_df.head(5)
```

```python
import re
def clean_tweet(tweet):
        '''
        Utility function to clean tweet text by removing links, special characters
        using simple regex statements.
        '''
        return (' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split()) )
```

```python
from textblob import TextBlob 
import pandas as pd 
```

```python
polarity = 0
positive = 0
negative = 0
neutral = 0
tweets_df['positive'] = 0
tweets_df['negative'] = 0
tweets_df['neutral'] = 0
for tweet in tweets_df['text']:
    
    cleaned_tweets = clean_tweet(tweet)
   
    analysis = TextBlob(cleaned_tweets)

    polarity += analysis.sentiment.polarity
    #print(polarity)
    if (analysis.sentiment.polarity == 0):  
        tweets_df['neutral'] += 1
    elif (analysis.sentiment.polarity > 0.0 ):
        tweets_df['positive'] += 1
    elif (analysis.sentiment.polarity < 0.0):
        tweets_df['negative'] += 1
    
del tweets_df['text']                
print(tweets_df.head(5))

```

```python
 # function to calculate percentage

def percentage(part, whole):
        temp = 100 * float(part) / float(whole)
        return format(temp, '.2f')
```

```python
positive = percentage(positive, 20)
negative = percentage(negative, 20)
neutral  = percentage(neutral, 20)
```
